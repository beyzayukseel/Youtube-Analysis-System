```python
df_['Bag_of_words'] = ''
columns = ['category', 'channel_name', 'descriptions']

for index, row in df_.iterrows():
    words = ''
    for col in columns:
        words += ' '.join(row[col]) + ' '
    row['Bag_of_words'] = words
    
    
df_['Bag_of_words'] = df_['Bag_of_words'].str.strip().str.replace('   ', '  ').str.replace('  ', ' ')

df_ = df_[['title_list','Bag_of_words']]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
import warnings
warnings.filterwarnings("ignore")
```

```python
count = CountVectorizer()
count_matrix = count.fit_transform(df_['Bag_of_words'])
count_matrix
```

```python
cosine_sim = cosine_similarity(count_matrix, count_matrix)
print(cosine_sim)
```

```python
indices = pd.Series(df_['title_list'])
indices[:5]
```

```python
def recommend(title, cosine_sim = cosine_sim):
    recommended_links = []
    idx = indices[indices == title].index[0]   
    score_series = pd.Series(cosine_sim[idx]).sort_values(ascending = False)   
    top_6_indices = list(score_series.iloc[1:6].index)   
    
    for i in top_6_indices:  
        recommended_links.append(list(df_['title_list'])[i])
        
    return recommended_links
```
